### Task 3: Calculating Total Loan Amount for Each Project

```python
#### In our analysis of World Bank loan data, the task is to determine the total original principal amount for all projects. 
#This calculation provides an essential snapshot of the combined financial commitment across all projects, offering insights 
#into their overall financial scale. This information is valuable for budgeting, resource allocation, and understanding the
#collective financial #impact of World Bank initiatives, aiding in informed decision-making and financial planning.
```

```python
%%sql
SELECT
    `Project name`,
    SUM(`Original Principal Amount`) AS total_loan_amount
FROM
    cleaned_dataset
GROUP BY
    `Project name`;
```

### Task 4: Calculating Total Original Principal Amount for All Projects

```python
%%sql
# a Common Table Expression (CTE) is used
With proj as
	(SELECT
    `Project name`,
    SUM(`Original Principal Amount`) AS total_loan_amount
FROM
    cleaned_dataset
GROUP BY
    `Project name`)
    select sum(total_loan_amount) as `Total Original Principal Amount` from proj;
```

### Task 5: Calculating Average Repaid to IDA for Each Region

```python
%%sql
SELECT
    DISTINCT Region, Round(AVG(`Repaid to IDA`),2) as Avg_Repaid
FROM 
    cleaned_dataset
GROUP BY 
    Region;
```

### Task 6: Identifying Country with Highest Repaid to IDA Ratio¶

```python
%%sql
#Calculate the max ratio of 'Repaid to IDA' to 'Original Principal Amount' for projects with Credit Status as 'Fully Repaid'
SELECT
    Country,
    MAX(`Repaid to IDA` / `Original Principal Amount`) AS MaxRepaidToPrincipalRatio
FROM
    cleaned_dataset
WHERE
    `Credit Status` = 'Fully Repaid'
GROUP BY
    Country;
```

### Task 7: Counting Projects by Credit Status for Each Country

```python
%%sql
# the aim is to  determine, for each country, 
#the number of projects with different credit status values, such as "Active," "Fully Repaid," "Cancelled," and others
SELECT 
		Country,  `Credit Status`, COUNT(`Project ID`) as NoOfProjects
FROM 
	cleaned_dataset
GROUP BY 
	Country, `Credit Status`
ORDER BY `NoOfProjects` DESC;
```

### Task 8: Counting Countries with World Bank Loans by Region

### Task 9: Counting Fully Repaid Projects by Region

```python
# TO calculate the total number of fully repaid projects (where "Credit Status" equals 'Fully Repaid') for each region
```

```python
%%sql
SELECT 
	Region, 
    SUM(CASE WHEN `Credit Status`= 'Fully Repaid' THEN 1 ELSE 0 END) as CountFullyRepaidProjects
FROM 
	cleaned_dataset
GROUP BY 
	Region;
```

### Task 10: Identifying Projects with the Highest "Due to IDA"

```python
#To search for projects with the highest "Due to IDA" amounts
```

```python
%%sql
SELECT
	Country,
	`Project Name`,
    `Effective Date`,
	CONCAT('$',Round(`Due to IDA`,2)) as `Due to IDA Amount`
FROM
	cleaned_dataset
ORDER BY `Due to IDA Amount` DESC
LIMIT 5;
```

### Task 11: Identifying Top 5 Countries with Highest Repaid to Principal Ratio

```python
%%sql
SELECT
    Country,
    CONCAT('$', ROUND(SUM(`Repaid to IDA`) / SUM(`Original Principal Amount`), 2), 'B') AS `Ratio Repaid to Original Principal Amount`
FROM
    cleaned_dataset
WHERE
    `Credit Status` <> 'Fully Repaid'
GROUP BY
    Country
ORDER BY
    `Ratio Repaid to Original Principal Amount` DESC
LIMIT 5;
```

### Task 12: Identifying Top 5 Countries with the Highest Total Loan Amount

```python
%%sql
SELECT
    country,
    CONCAT('$', ROUND(SUM(CAST(REPLACE(`Original Principal Amount`, ',', '') AS DECIMAL(10, 2)))
                    - SUM(CAST(REPLACE(`Cancelled Amount`, ',', '') AS DECIMAL(10, 2))) / 1000000000, 2), 'B') AS 'All Loan Amount'
FROM
    cleaned_dataset
GROUP BY
    country
ORDER BY
    'All Loan Amount' DESC
LIMIT 5;
```

### Task 13: Identifying Top 5 Countries with the Highest Due Amount

```python
%%sql
SELECT
    country,
    CONCAT('$', ROUND(SUM(CAST(REPLACE(`Due to IDA`, ',', '') AS DECIMAL(10, 2))) / 1000000000, 2), 'B') AS 'Due Loan Amount'
FROM
    cleaned_dataset
GROUP BY
    country
ORDER BY
    SUM(CAST(REPLACE(`Due to IDA`, ',', '') AS DECIMAL(10, 2))) / 1000000 DESC
LIMIT 5;
```
